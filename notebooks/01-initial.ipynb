# CMI - Detect Behavior With Sensor Data -- Initial Exploration

## Key Points From the Overview

### **Competition Details**

**The goal of this competition is to develop a predictive model that distinguishes BFRB-like and non-BFRB-like activity using data from a variety of sensors collected via a wrist-worn device. Successfully disentangling these behaviors will improve the design and accuracy of wearable BFRB-detection devices, which are relevant to a wide range of mental illnesses, ultimately strengthening the tools available to support their treatment.**

- To investigate BFRBs, the Child Mind Institute has developed a wrist-worn device, Helios, designed to detect these behaviors. While many commercially available devices contain Inertial Measurement Units (IMUs) to measure rotation and motion, the Helios watch integrates additional sensors, including 5 thermopiles (for detecting body heat) and 5 time-of-flight sensors (for detecting proximity). See the figure to the right for the placement of these sensors on the Helios device.

#### `Experiment`
- They began a transition from “rest” position and moved their hand to the appropriate location (Transition);
- They followed this with a short pause wherein they did nothing (Pause); and
- Finally they performed a gesture from either the BFRB-like or non-BFRB-like category of movements.
- Each participant performed 18 unique gestures (8 BFRB-like gestures and 10 non-BFRB-like gestures) in at least 1 of 4 different body-positions (sitting, sitting leaning forward with their non-dominant arm resting on their leg, lying on their back, and lying on their side).

#### `Challenge`

- This competition challenges you to develop a predictive model capable of distinguishing (1) BFRB-like gestures from non-BFRB-like gestures and (2) the specific type of BFRB-like gesture. 
- **Critically, when your model is evaluated, half of the test set will include only data from the IMU, while the other half will include all of the sensors on the Helios device (IMU, thermopiles, and time-of-flight sensors).**
- Your solutions will have direct real-world impact, as the insights gained will inform design decisions about sensor selection — specifically whether the added expense and complexity of thermopile and time-of-flight sensors is justified by significant improvements in BFRB detection accuracy compared to an IMU alone. 
    - By helping us determine the added value of these thermopiles and time-of-flight sensors, your work will guide the development of better tools for detection and treatment of BFRBs.


### **Data**

**`Types of Gestures`** 

- 8 BFRB
- 10 non- BFRB gestures

**`Helios Sensors + Hidden Test Set Info`**

- 1x Inertial Measurement Unit (IMU; BNO080/BNO085): An integrated sensor that combines accelerometer, gyroscope, and magnetometer measurements with onboard processing to provide orientation and motion data.
- 5x Thermopile Sensor (MLX90632): A non-contact temperature sensor that measures infrared radiation.
- 5x Time-of-Flight Sensor (VL53L7CX): A sensor that measures distance by detecting how long it takes for emitted infrared light to bounce back from objects.
- 3,500 sequences in the hidden test set.
- **Half of the hidden-test sequences are recorded with IMU only; the thermopile (thm_) and time-of-flight (tof__v*) columns are still present but contain null values for those sequences.**

**`Features / Columns to be Mindfull Of`**

- **sequence_type** target or non target gesture - train only.
- **gesture** `target column` - description of sequence gesture. train only.
- **orientation** subject orientation during sequence - train only.

```python
## Move to utils
def download_data(competition:str, base_path: Path):
    data_path = base_path/'data'
    zip_path = data_path / f"{competition}.zip"
    
    data_path.mkdir(parents=True, exist_ok=False)
    print(f"Downloading data for competition: {competition}...")
    try:
        command = [
            "kaggle", "competitions", "download",
            "-c", competition,
            "-p", str(data_path),
            "--force"
            ]
        result = subprocess.run(
            command,
            check=True,
            text=True,
            encoding='utf-8'
        )
        print("Kaggle API Output: ", result.stdout)
    except FileNotFoundError: 
        print("'kaggle' command not found")
        return
    except subprocess.CalledProcessError as e:
        print(f"Error downloading data. Kaggle returned an API error: {e}")
        return

    try:
        with zipfile.ZipFile(zip_path, 'r') as z:
            z.extractall(data_path)
        print("Unzipping Complete.")
    except zipfile.BadZipFile:
        print(f"Error: Failed to unzip {zip_path}. The file may be corrupted.")
        return
    
    os.remove(zip_path)
        
```

#### Initial takeaways

- With 81 individual subjects and 18 unique gestures per subject, this is a non-trivial multi-class problem.
- Looking at NaN values per column, there appear to be two distinct types of sensor failures:
    - **Intermittent Failures** `rot_*` [quarternions](https://en.wikipedia.org/wiki/Quaternion), `thm_1-4` and `tof_1-4` sensors exhibit low and consistent NaN percentages (~0.64 and ~1.08%). Could this possibly be because of sporadic, temporary loss of communication from specific sensor chips?
    - **Systematic Failure** `thm_5` and all of `tof_5` pixels have a much higher and identical failure rate (~5.5%). Suggesting, quite strongly, that the fifth group of sensors is less reliable or could have been partially disabled on some devices.
- For many of the `tof_*` pixels the 25th, and sometimes the 50th, percentiles are -1.
    - This indicates that -1 is not an outlier but a frequent state representing "no object detected in range". This is fundamental to the data.
    - The value must be treated carefully. During feature engineering statistics need to be calculated for both with -1 and without -1 values.
- Moving to categorical variables, the `phase` column (Transition, Pause and Gesture) segments the behaviour within a sequence.

```python
# Extended visualizations
def visualize_sequence(df, sequence_id):
    """Visualize all sensor data for a single sequence"""
    seq_data = df[df["sequence_id"] == sequence_id]
    fig, axes = plt.subplots(3, 1, figsize=(10, 8))
    
    # Thermopile
    thm_cols = [col for col in df.columns if 'thm_' in col]
    for col in thm_cols:
        sns.lineplot(data=seq_data, x='sequence_counter', y=col, label=col, ax=axes[0])
    axes[0].set_title('Thermopile Data')
    axes[0].set_xlabel('Value')

    # Time of Flight
    tof_sensors_mean = {}
    tof_sensors_med = {}
    for i in range(1, 6):
        sensor_cols = [f'tof_{i}_v{j}' for j in range(64)]
        tof_sensors_mean[f'tof_sensor{i}'] = seq_data[sensor_cols].mean(axis=1)
        tof_sensors_med[f'tof_sensor{i}'] = seq_data[sensor_cols].median(axis=1)
    
    for sensor, values in tof_sensors_mean.items():
        sns.lineplot(x=seq_data['sequence_counter'], y=values, ax=axes[1], label=sensor)
    axes[1].set_title('Time-of-Flight Data (Sensor-wise Average)')
    axes[1].set_xlabel('Value')
    
    for sensor, values in tof_sensors_med.items():
        sns.lineplot(x=seq_data['sequence_counter'], y=values, ax=axes[2], label=sensor)
    axes[2].set_title('Time-of-Flight Data (Sensor-wise Median)')
    axes[2].set_xlabel('Value')
    
    plt.tight_layout()
    plt.savefig('imgs/visual-sequence')
    plt.show();
```

```python
def plot_single_sequence(df, sequence_id):
    seq_df = df[df['sequence_id'] == sequence_id]
    
    # IMU Sensors (acc_, rot_)
    imu_cols = ['acc_x', 'acc_y', 'acc_z', 'rot_w', 'rot_x', 'rot_y', 'rot_z']
    plt.figure(figsize=(14, 6))
    for col in imu_cols:
        plt.plot(seq_df['sequence_counter'], seq_df[col], label=col)
    plt.title(f"IMU Sensors for Sequence {sequence_id}")
    plt.xlabel("Sequence Counter")
    plt.ylabel("Sensor Value")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # Thermopile Sensors (thm_1 to thm_5)
    thm_cols = [f'thm_{i}' for i in range(1, 6)]
    plt.figure(figsize=(14, 4))
    for col in thm_cols:
        plt.plot(seq_df['sequence_counter'], seq_df[col], label=col)
    plt.title(f"Thermopile Sensors for Sequence {sequence_id}")
    plt.xlabel("Sequence Counter")
    plt.ylabel("Temperature (°C)")
    plt.legend()
    plt.grid(True)
    plt.show()
    
    # ToF Sensors (tof_*_v*)
    tof_cols = [c for c in df.columns if c.startswith('tof_') and '_v' in c]
    tof_sensors = sorted(set(c.split('_')[1] for c in tof_cols))  # Get unique ToF sensor indices
    
    for sensor_idx in tof_sensors:
        sensor_pixels = [c for c in tof_cols if c.startswith(f'tof_{sensor_idx}_v')]
        pixel_data = seq_df[sensor_pixels].values[0]  # First row of sequence
        
        plt.figure(figsize=(8, 6))
        plt.imshow(pixel_data.reshape(8, 8), cmap='viridis', interpolation='nearest')
        plt.colorbar(label='Uncalibrated Value')
        plt.title(f"ToF Sensor {sensor_idx} Grid for Sequence {sequence_id}")
        plt.axis('off')
        plt.show();
```

Visualizing spatial information of ToF sensors at key timesteps. This visualizes the 8x8 grid for each of the 5 ToF sensors at one point in time from each phase.

```python
# Helper function to reshape ToF
def reshape_tof_pixels(row, sensor_index):
    # -1 replaced with NaN
    tof_pixels = row[[f'tof_{sensor_index}_v{i}' for i in range(64)]].values.astype(float)
    tof_pixels[tof_pixels == -1] = np.nan
    return tof_pixels.reshape(8, 8)
```

```python
phases = ['Transition', 'Pause', 'Gesture']
timesteps = {}

for p in phases:
    phase_counters = sequence_data[sequence_data['phase'] == p]['sequence_counter']
    if not phase_counters.empty:
        timesteps[p] = phase_counters.iloc[len(phase_counters) //  2]
        
if timesteps:
    fig, axes = plt.subplots(len(timesteps), 5, figsize=(15, 3 * len(timesteps) + 1),
                             gridspec_kw={'width_ratios': [1, 1, 1, 1, 1]})
    fig.suptitle(f'ToF Sensor Grids at Key Timesteps for Sequence: {example_sequence_id}', fontsize=16, y=1.02)
    
    for i, (phase, timestep) in enumerate(timesteps.items()):
        row_data = sequence_data[sequence_data['sequence_counter'] == timestep]
        for j in range(1, 6):
           ax = axes[i, j-1]
           if not row_data.empty:
               grid = reshape_tof_pixels(row_data.iloc[0], j)
               # Heatmap with fixed color scale
               sns.heatmap(grid, ax=ax, cbar=i==0 and j==5, cmap='viridis', annot=False,
                           vmin=0, vmax=254)
           ax.set_title(f"ToF {j} @ {phase} (t={timestep})")
           ax.set_xticks([])
           ax.set_yticks([])

    plt.tight_layout()
    plt.show();
else:
    print("Couldn't find all phases in the example sequence to generate ToF heatmaps.") 
               
    
```

```python
def compute_sensor_statistics(df, sensor_groups):
    stats = {}
    for group_name, cols in sensor_groups.items():
        group_df = df[cols]
        stats[group_name] = {
            'mean': group_df.mean(),
            'median': group_df.median(),
            'iqr': group_df.quantile(0.75) - group_df.quantile(0.25)
        }
    return pd.DataFrame(stats)

# Define sensor groups
sensor_groups = {
    'IMU_Acceleration': ['acc_x', 'acc_y', 'acc_z'],
    'IMU_Orientation': ['rot_w', 'rot_x', 'rot_y', 'rot_z'],
    'Thermopile': [f'thm_{i}' for i in range(1, 6)],
    'ToF_Sensor1': [f'tof_1_v{i}' for i in range(64)],
    'ToF_Sensor2': [f'tof_2_v{i}' for i in range(64)],
    # Add other ToF sensors if needed
}

# Compute statistics
sensor_stats = compute_sensor_statistics(train_full, sensor_groups)
display_all(sensor_stats)
```

```python
def analyze_by_behavior_phase(df):
    phases = df['behavior'].unique()
    phase_stats = {}
    
    for phase in phases:
        phase_df = df[df['behavior'] == phase]
        phase_stats[phase] = {
            'IMU_Acceleration_Mean': phase_df[['acc_x', 'acc_y', 'acc_z']].mean().to_dict(),
            'Thermopile_Range': {
                col: (phase_df[col].min(), phase_df[col].max()) 
                for col in [f'thm_{i}' for i in range(1, 6)]
            },
            'ToF_Sensor1_Missing': (phase_df['tof_1_v0'] == -1).mean()  # % of missing ToF values
        }
    
    return phase_stats

# Example usage
phase_analysis = analyze_by_behavior_phase(train_full)
print(phase_analysis)
```

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
```

```python
# Modular EDA functions
def plot_target_distribution(df):
    """Analyzes and plots the distribution of the target variable, grouped by type"""
    
    # Gesture counts grouped by sequence types
    gest_cnt = df.groupby(['gesture', 'sequence_type']).size().reset_index(name='count')
    
    # Order of gestures by total frequency
    gest_ord = df['gesture'].value_counts().index
    
    fig = px.bar(
        gest_cnt,
        x='gesture',
        y='count',
        color='sequence_type',
        title='Distribution of Gestures (BFRB vs. Non-BFRB)',
        category_orders={'gesture': gest_ord},
        labels={'gesture': 'Gesture', 'count': 'Number of Readings', 'sequence_type': ' Sequence Type'},
        text_auto=True
    )
    fig.update_layout(xaxis_tickangle=-45)
    fig.show();
    

def plot_demographic_distribution(df):
    """Plots distributions of key demographic faetures"""
    
    demo_df = df.drop_duplicates(subset=['subject'])
    
    fig = make_subplots(
        rows=1, cols=3,
        subplot_titles=('Sex Distribution', 'Adult/Child Distribution', 'Handedness Distribution')
    )
    
    sex_cnts = demo_df['sex'].map({0: 'Female', 1: 'Male'}).value_counts()
    adult_child_cnts = demo_df['adult_child'].map({0: 'Child', 1: 'Adult'}).value_counts()
    handedness_cnts = demo_df['handedness'].map({0: 'Left', 1: 'Right'}).value_counts()
    
    fig.add_trace(go.Bar(x=sex_cnts.index, y=sex_cnts.values, name='Sex'), row=1, col=1)
    fig.add_trace(go.Bar(x=adult_child_cnts.index, y=adult_child_cnts.values, name='Age Group'), row=1, col=2)   
    fig.add_trace(go.Bar(x=handedness_cnts.index, y=handedness_cnts.values, name='Sex'), row=1, col=3)
    
    fig.update_layout(title_text='Demographic Feature Distributions', showlegend=False)
    fig.show();       


def visualize_full_sequence(df, sequence_id):
    """Create sensor specifc visualizations for a single sequence, highlighting phases.
    - IMU: Interactive line plot with calculated magnitude.
    - Thermopile: Interactive line Plot of temperature readings.
    - ToF: Heatmap of sensor activations over time for each phase.
    """
    sequence_data = df[df['sequence_id'] == sequence_id].copy()
    sequence_data.sort_values('sequence_counter', inplace=True)
    
    # IMU visualization (with magnitude and jerk)
    acc_cols = ['acc_x', 'acc_y', 'acc_z']
    rot_cols = ['rot_w', 'rot_x', 'rot_y', 'rot_z']
    
    # Calculate magnitude and jerk
    sequence_data['acc_mag'] = np.sqrt((sequence_data[acc_cols]**2).sum(axis=1))
    sequence_data['jerk'] = sequence_data['acc_mag'].diff().fillna(0)
    
    imu_plot_cols = acc_cols + ['acc_mag', 'jerk'] + rot_cols
    
    fig_imu = px.line(
        sequence_data,
        x='sequence_counter',
        y=imu_plot_cols,
        color='phase',
        facet_row='variable',
        title=f'IMU Data for Sequence {sequence_id}',
        labels={'value': 'Sensor Reading', 'sequence_counter': 'Time Step'},
        height=1200
    )
    fig_imu.update_yaxes(matches=None)
    fig_imu.show()
    
    # Thermopile visualization
    thm_cols = [f'thm_{i}' for i in range(1, 6)]
    fig_thm = px.line(
        sequence_data,
        x='sequence_counter',
        y=thm_cols,
        color='phase',
        title=f'Thermopile Data for Sequence {sequence_id}',
        labels={'value': 'Temperature (C)', 'sequence_counter': 'Time Step'},
    )
    fig_thm.show()
    
    # ToF visualization with phase specific heatmaps
    tof_cols = [f'tof_{s}_v{p}' for s in range(1, 6) for p in range(64)]
    
    # Replace -1 with NaN for visualization 
    sequence_data[tof_cols] = sequence_data[tof_cols].replace(-1, np.nan)
    
    phases = sequence_data['phase'].unique()
    fig, axes = plt.subplots(len(phases), 1, figsize=(20, 8 * len(phases)), sharex=True)
    if len(phases) == 1: axes = [axes] # Ensure axes is iterable for a single phase
    
    fig.suptitle(f"ToF Sensor Activation Heatmaps for Sequence {sequence_id}", fontsize=20, y=0.95)
    
    for i, phase in enumerate(phases):
        phase_data = sequence_data[sequence_data['phase'] == phase]
        
        # Reshape ToF data for the heatmap
        tof_pixel_data = phase_data[tof_cols].T
        
        sns.heatmap(tof_pixel_data, ax=axes[i], cmap='viridis', cbar=True)
        axes[i].set_title(f'Phase: {phase}', fontsize=16)
        axes[i].set_xlabel('Sequence Counter (Time Step)')
        axes[i].set_ylabel('ToF Sensor Pixel (0-319)')
        
    plt.tight_layout(rect=[0, 0, 1, 0.93])
    plt.show();
```
